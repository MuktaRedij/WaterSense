```python
import pandas as pd

df = pd.read_csv("C:\\Users\\Mukta\\Downloads\\district_groundwater.csv")  # rename based on actual file name
df.columns = df.columns.str.strip()   # clean column names
df.head()
```

```python
def risk_level(value):
    if value >= 90:
        return "ðŸ”´ Severe Risk"
    elif value >= 75:
        return "ðŸŸ  High Risk"
    elif value >= 50:
        return "ðŸŸ¡ Medium Risk"
    else:
        return "ðŸŸ¢ Low Risk"

df["Risk Level"] = df[EXTRACTION_COL].apply(risk_level)
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import matplotlib.pyplot as plt

EXTRACTION_COL = "Stage of Ground Water Extraction (%)"

districts = sorted(df["Name of District"].dropna().unique())

district_dropdown = widgets.Dropdown(
    options=districts,
    description='District:',
    style={'description_width': '80px'},
    layout=widgets.Layout(width='350px')
)

predict_button = widgets.Button(
    description='Predict Groundwater Status',
    button_style='primary',
    layout=widgets.Layout(width='300px')
)

output = widgets.Output()

def on_predict_clicked(b):
    with output:
        clear_output()
        selected = district_dropdown.value
        row = df[df["Name of District"] == selected].iloc[0]

        extraction = row[EXTRACTION_COL]
        risk = row["Risk Level"]
        state = row["Name of State"]

        print("ðŸ’§ WaterSense â€“ Prototype Result")
        print("------------------------------------")
        print(f"ðŸ“ Location : {selected}, {state}")
        print(f"ðŸ’¦ Extraction: {extraction:.1f}%")
        print(f"âš  Risk      : {risk}")
        print()

        if "Severe" in risk:
            print("â— Critical water stress â€“ avoid new borewells, drip irrigation required.")
        elif "High" in risk:
            print("âš  High pressure â€“ regulated usage and monitoring advised.")
        elif "Medium" in risk:
            print("ðŸŸ¡ Moderate risk â€“ controlled irrigation recommended.")
        else:
            print("ðŸŸ¢ Stable â€“ normal usage possible.")

        # Bar visualization
        plt.figure(figsize=(5,2))
        color = "red" if extraction >= 90 else "orange" if extraction >= 75 else "yellow" if extraction >= 50 else "green"
        plt.barh([selected], [extraction], color=color)
        plt.xlim(0,120)
        plt.xlabel("Groundwater Extraction (%)")
        plt.title("Groundwater Status")
        plt.show()

predict_button.on_click(on_predict_clicked)

ui = widgets.VBox([
    widgets.HTML("<h3>ðŸ’§ WaterSense â€“ Groundwater Risk Demo</h3>"),
    district_dropdown,
    predict_button,
    output
])

display(ui)
```

```python
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# Load Dataset
# -----------------------------

df = pd.read_csv("C:\\Users\\Mukta\\Downloads\\district_groundwater.csv")
df.columns = df.columns.str.strip()

# Use the real column names:
DIST_COL = "Name of District"
STATE_COL = "Name of State"
EXTRACTION_COL = "Stage of Ground Water Extraction (%)"

# -----------------------------
# Risk Classification
# -----------------------------

def risk_level(value):
    if value >= 90:
        return "ðŸ”´ Severe Risk"
    elif value >= 75:
        return "ðŸŸ  High Risk"
    elif value >= 50:
        return "ðŸŸ¡ Medium Risk"
    else:
        return "ðŸŸ¢ Low Risk"

df["Risk Level"] = df[EXTRACTION_COL].apply(risk_level)

# -----------------------------
# UI SECTION
# -----------------------------

# Title
st.title("ðŸ’§ WaterSense - Groundwater Risk Prototype")

st.write("Select a district to view groundwater risk assessment based on national dataset.")

# Dropdown selection
district = st.selectbox("Choose District:", sorted(df[DIST_COL].unique()))

# Fetch row
row = df[df[DIST_COL] == district].iloc[0]

# Display result
st.subheader(f"ðŸ“ {district}, {row[STATE_COL]}")

extraction = row[EXTRACTION_COL]
risk = row["Risk Level"]

st.write(f"### ðŸ’¦ Groundwater Extraction: **{extraction:.1f}%**")
st.write(f"### âš  Risk Level: **{risk}**")

# Recommendation logic
if "Severe" in risk:
    suggestion = "â— Critical water stress â€” avoid drilling new borewells, mandatory drip irrigation."
elif "High" in risk:
    suggestion = "âš  High pressure â€” regulated usage and monitoring advised."
elif "Medium" in risk:
    suggestion = "ðŸŸ¡ Moderate risk â€” controlled irrigation recommended."
else:
    suggestion = "ðŸŸ¢ Safe usage â€” normal irrigation is acceptable with periodic monitoring."

st.write("### ðŸ“ Recommendation:")
st.success(suggestion)

# -----------------------------
# Chart
# -----------------------------

st.write("### ðŸ“Š Groundwater Stress Visualization")

fig, ax = plt.subplots(figsize=(5,2))

color = "red" if extraction >= 90 else "orange" if extraction >= 75 else "gold" if extraction >= 50 else "green"
ax.barh([district], [extraction], color=color)
ax.set_xlim(0, 120)
ax.set_xlabel("Groundwater Extraction (%)")

st.pyplot(fig)
```
